```python
def tweets_scraper(username, file_path):
    # Configure
    c = twint.Config()
    
    # Twitter user's username
    c.Username = username
    
    # Filter Tweets sent since date, works only with twint.run.Search (Example: 2017-12-27).
    c.Since = '2020-01-01'
    # Filter Tweets sent until date, works only with twint.run.Search (Example: 2017-12-27).
    #c.Until = '2020-10-19'
    
    # Name of the output file.
    c.Output = file_path
    
    # Set to True to write as a csv file.
    c.Tabs = True
    c.Store_csv = True
    
    # Hide output.
    c.Hide_output = True
    
    # Run
    twint.run.Search(c)
```

```python
def retweets_scraper(username, file_path):
    # Configure
    c = twint.Config()
    
    # Twitter user's username
    c.Username = username
    
    # Filter Tweets sent since date, works only with twint.run.Search (Example: 2017-12-27).
    c.Since = '2020-09-19'
    # Filter Tweets sent until date, works only with twint.run.Search (Example: 2017-12-27).
    #c.Until = '2020-10-19'
    
    # Get retweets done by the user.
    #c.Profile_full = True
    #c.Retweets = True    # doesn't work
    #c.Filter_retweets = False
    
    # Name of the output file.
    c.Output = file_path
    
    # Set to True to write as a csv file.
    c.Tabs = True
    c.Store_csv = True
    
    # Hide output.
    c.Hide_output = True
    
    # Run
    twint.run.Search(c)
```

```python
def replies_scraper(username, file_path):
    # Configure
    c = twint.Config()
    
    # Filter Tweets sent since date, works only with twint.run.Search (Example: 2017-12-27).
    c.Since = '2020-09-19'
    # Filter Tweets sent until date, works only with twint.run.Search (Example: 2017-12-27).
    #c.Until = '2020-10-19'
    
    # Display Tweets tweeted _to_ the specified user
    c.To = username
    
    # Name of the output file.
    c.Output = file_path
    
    # Set to True to write as a csv file.
    c.Tabs = True
    c.Store_csv = True
    
    # Hide output.
    c.Hide_output = True
    
    # Run
    twint.run.Search(c)
```

### PREPARE DATA FOR TASK 2A

```python
counts_df_list = []

for i, df in enumerate(df_list):
    counts_df_list.append(df['date'].value_counts().rename_axis('date').reset_index(name='counts'))
    print(i, ' - ', counts_df_list[i].shape)
    
#counts_df_list[0]
```

```python
counts_df = reduce(lambda left,right: pd.merge(left, right, on='date', how='outer'), counts_df_list)

parties_names = []
for df in df_list:
    parties_names.append(df['name'].iloc[0])

cols = ['date']
cols.extend(parties_names)

counts_df.columns = cols

for party in parties_names:
    counts_df = counts_df.fillna(0).astype({party: np.int64})

counts_df.sort_values(by=['date'], inplace=True, ignore_index=True)
counts_df
```
