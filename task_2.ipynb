### TASK 2C

```python
filelist = []
filename_part = 'out/2c/activity_'

for party in parties_flattened:
    filelist.append(filename_part + party + '.csv')

df_list = [pd.read_csv(file) for file in filelist]

parties_names = []
for i, df in enumerate(df_list):
    parties_names.append(df['name'].iloc[0])
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    print(i, ' - ', df.shape)
    
df_list[0]
```

```python
fig, ax = plt.subplots(figsize=(16, 8))

for i, df in enumerate(df_list):
    df = df[(df.timestamp >= datetime.datetime(2020, 10, 1)) & 
            (df.timestamp <= datetime.datetime(2020, 10, 8))]
    
    grp = df.groupby(pd.Grouper(key='timestamp', freq='D'))['id'].count()
    
    if grp.shape != (0,):
        ax = grp.plot(ax=ax, kind='line', label=parties_names[i], style='-')
    else:
        print('NOTE: ' + parties_names[i] + ' did not tweet at the given time.')
        
plt.xlabel('Date')
plt.ylabel('Num of tweets')
plt.legend(loc='upper right')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 8))

for i, df in enumerate(df_list):
    df = df[(df.timestamp >= datetime.datetime(2020, 10, 1)) & 
            (df.timestamp <= datetime.datetime(2020, 10, 2))]
    
    grp = df.groupby(pd.Grouper(key='timestamp', freq='H'))['id'].count()
    
    if grp.shape != (0,):
        ax = grp.plot(ax=ax, kind='line', label=parties_names[i], style='-')
    else:
        print('NOTE: ' + parties_names[i] + ' did not tweet at the given time.')
        
plt.xlabel('Time')
plt.ylabel('Num of tweets')
plt.legend(loc='upper right')
plt.show()
```
